```python
import numpy as np
import matplotlib.pyplot as plt
```

1.

```python
firstCol = np.ones(shape=[5,1])
secondCol = np.arange(1,6).reshape(5,1)
thirdCol = (np.arange(1,6).reshape(5,1))**2
fourthCol = (np.arange(1,6).reshape(5,1))**3
fifthCol = (np.arange(1,6).reshape(5,1))**4
X = np.concatenate((firstCol, secondCol, thirdCol, fourthCol, fifthCol), axis=1)
print(X)
```

2.

```python
x = np.arange(1,11).reshape(10,1)
y = np.array([14, 12, 34, 24, 43, 38, 41, 57, 55, 50]).reshape(10,1)
ones = np.ones(10).reshape(10,1)
X = np.concatenate((ones,x), axis=1)
xx = np.linspace(1, 10, 1000)

XX = np.concatenate((np.ones(1000).reshape(1000,1), xx.reshape(1000,1)), axis= 1)
M = np.linalg.inv(np.dot(X.T, X)).dot(X.T).dot(y)
yy = XX.dot(M)
plt.plot(x, y, 'o')
plt.plot(xx, yy)
plt.show()
```

3.

```python
X = np.concatenate((X, x**2), axis=1)
M = np.linalg.inv(np.dot(X.T, X)).dot(X.T).dot(y)
xx = np.linspace(1, 10, 1000)
XX = np.concatenate((np.ones((1000,1)), xx.reshape(1000,1), xx.reshape(1000,1)**2), axis=1)

yy = XX.dot(M)
plt.plot(xx, yy)
plt.plot(x, y, 'o')
plt.show()
```

4.

```python
x = np.arange(1,11)
y = np.array([14, 12, 34, 24, 43, 38, 41, 57, 55, 50])
coef = np.polyfit(x, y, 3)
p = np.poly1d(coef)
xx = np.linspace(1, 10, 1000)
yy = p(xx)
plt.plot(x, y, 'o')
plt.plot(xx, yy)
plt.show()
```

When using higher polynomials predictions starts to overfit. I would use 2 or 3 polynomial for prediction.
