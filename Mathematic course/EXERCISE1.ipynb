```python
x = np.linspace(-2.5,2.5, 10000)
f = (x**3 + np.abs(x) + np.log(x + 10**-8)) / (np.sqrt(100 + x) + 0.01 * x * np.exp(x))
plt.plot(x,f)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('awesome graph')
plt.show()
```
